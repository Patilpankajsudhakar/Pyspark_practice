```python
# no header in the dataset let suppose
data = [(1,"John","CEO","10000")
        ,(2,"Mary","CTO","12000")
        ,(3,"Mike","CFO","15000")
        ,(4,"Sara","CIO","18000")
        ,(5,"Tom","COO","20000")]

# option providing schema our self
schema_str = "id INT, name STRING, title STRING, salary STRING"

df = spark.createDataFrame(data, schema=schema_str)
df.printSchema()
df.show()
#
```

```python
# create the dataframe and create schema
from pyspark.sql.types import StructType, StructField, StringType, DoubleType, IntegerType
schema = StructType([
    StructField("id", IntegerType(), True),
    StructField('name',StringType(),True),
    StructField("title",StringType(),True),
    StructField("salary",StringType(),True)
])
df = spark.createDataFrame(data,schema)
df.printSchema()
df.show()
```
